```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
#import os
```

```python
Rlineal = pd.read_csv("C:/Users/Pablo Álvarez A/Pablo Alvarez/Regresión lineal/data.csv")#llamamos el archivo.
Rlineal.head(5)
```

```python
figure = plt.figure(figsize=(8,8))
plt.scatter(Rlineal['precio'], Rlineal['metro'])
plt.plot(Rlineal['precio'], Rlineal['metro'])
plt.xlabel('precio')
plt.ylabel('Metro')
plt.grid()
```

```python
Rlineal.keys()
```

```python
Iprecio = Rlineal['precio'].values.reshape(-1,1)
Imetro = Rlineal['metro'].values.reshape(-1,1)
linear_regressor = LinearRegression()
linear_regressor.fit(Iprecio,Imetro)
Imetro_pred = linear_regressor.predict(Iprecio)
Imetro_pred
```

```python
# y = mx +c formula
m = linear_regressor.coef_[0][0]
c = linear_regressor.intercept_[0]
etiqueta = r'$metros = %5f * precio %+4f$'%(m,c)
print(etiqueta)
```

```python
figure = plt.figure(figsize=(8,8))
#plt.scatter(Rlineal['precio'], Rlineal['metro'])
plt.plot(Rlineal['precio'], Rlineal['metro'], label = 'precio')
plt.plot(Iprecio,Imetro_pred, color = 'red', label = etiqueta)
plt.xlabel('precio')
plt.ylabel('Metro')
plt.legend()
plt.grid()
```
